```python
import random

# функция загрузки файла
def load_from_file(file):
    f = open(file, encoding="utf-8")
    data = [line.strip() for line in f]
    f.close()
    
    return data

# запись в файл
def write_in_file(file, data_list):
    f = open(file, 'w', encoding="utf-8")
    try:
        for data in data_list:
            f.write(data + '\n')
    except:
        f.write('При записи произошла ошибка' + '\n')
    f.close()
    
    return True

# Очистка истории
def clear_history(file):
    f = open(file)
    f.seek(0)
    f.close()
    
    return True


#Поиск посследней буквы города
def seach_last_letter(city):
    if city[-1].upper() == 'Ъ' or city[-1].upper() == 'Ь':
        return city[-2].upper()
    else:
        return city[-1].upper()
    

# Создание алфавита для словаря
def create_alphabet():
    new_list = []
    for i in range(1040, 1071):
        new_list.append(chr(i))
    
    return new_list
        
#Создание словаря с городами
def create_dict(city_list):
    '''Из списка городов (из файла) и списка с алфавитом создается словарь, где ключи - буквы алфавита,
а значение - города на эту букву '''
    alphabet = create_alphabet()
    new_dict = {}
    
    for city in city_list:
        city = city
    
    city_list.sort()
    
    
    for i in alphabet:
        temp_list = []
        for city in city_list:
            if city[0] == i:
                temp_list.append(city)
        new_dict[i] = temp_list
        
    return new_dict

# Выбор нужного города из списка
def matching(letter):
    '''Берестся список городов на нужную букву и выбирается рандомно город, если город уже использовался ранее, выбирается другой'''
    need_cityes = cityes[letter]
    need_city = []
    
    for city in need_cityes:
        if city not in history:
            need_city.append(city)
            
    if len(need_city) > 0:       
        city = need_city[random.randint(0, len(need_city) - 1)]
        history.append(city)
        print(f'{city}. Вам на {seach_last_letter(city)}!')
    
        return seach_last_letter(city), True
    else:
        print('Вы победили!')
        processing = False
        return False, False
        
    
# глобальные перменный города и ответы

city_list = list(set(load_from_file('city_list.txt')))
cityes = create_dict(city_list)
history = load_from_file('answer.txt')
    
#Запуск игры    
def app():
    processing = True
    letter = 'a'
    count = 0
    
    #печатаем инструкцию при открытии
    data = load_from_file('hotKey.txt')
    print(*data, sep = '\n')
    print()
    
    
    while processing:
        answer = input('Введите слово или ключевое слово: ')
        
        answer = answer.strip().title()
        
        #Проверка, была ли запущена раньше игра: проверяем историю
        if len(history) > 0:
            count += 1
            letter = seach_last_letter(history[-1])
            
        
        
        #Основной роутинг
        if answer == 'About':
            data = load_from_file('describtion.txt')
            print(*data, sep = '\n')
            print()
        elif answer == 'Help':
            data = load_from_file('hotKey.txt')
            print(*data, sep = '\n')
            print()
        elif answer == 'History':
            write_in_file('answer.txt', history)
            data = load_from_file('answer.txt')
            print(*data, sep = '\n')
            print()
        elif answer == 'Clear':
            clear_history('answer.txt')
            history.clear()
            print('История очищена!')
            print()
        elif answer == 'Exit':
            write_in_file('answer.txt', history)
            print('Игра прервана!')
            processing = False
            break
        elif ord(answer[0].upper()) < 1040 or ord(answer[0].upper()) > 1071:
            print('Введите город на русском языке!')
            print()
        elif len(answer) < 2:
            print('Такого города нет!')
            print()
        elif answer[0] == letter or count == 0:
            if answer in history:
                if count == 0:
                    print(f'Данный город уже был!')
                else:    
                    print(f'Данный город уже был. Вам на {letter}!')
            else:
                history.append(answer)
                new_letter = seach_last_letter(answer)
                letter,processing = matching(new_letter)
                if count == 0:
                    count += 1
        else:
            print(f'Вы ввели город не на ту букву, Вам на {letter}!')
                
            
        

app()
```
